```python
import pandas as pd

medals = pd.read_csv('/content/drive/MyDrive/DATA/olympic_medals.csv')

medals.info()
```

```python
# (a) Find the size of the data frame. How many rows, how many columns?
num_rows, num_cols = medals.shape
print("Number of rows: {num_rows}, Number of columns: {num_cols}")
```

```python
# (b) Use the appropriate function to report the data types for all columns.
print(medals.dtypes)
```

```python
# (c) Find how many unique cities matches were held at
len(medals['Location'].unique())
```

```python
# (d) Find how many medals the USA team won in total.
usa_medals = len(medals[medals['Nationality'] == 'USA'])
print("Total medals won by the USA team:" , usa_medals)
```

```python
# (e) Find the total number of medals for each Nationality

medals.groupby('Nationality')['Medal'].count()
```
